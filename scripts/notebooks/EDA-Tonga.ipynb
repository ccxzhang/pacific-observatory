```python
import os 
os.chdir("../..")
import numpy
import numpy as np
import pandas as pd
import datetime
from scripts.PdfParse import *
```

```python
tonga_folder = os.getcwd() + "/data/tourism/tonga/temp/"
tonga_files = os.listdir(tonga_folder)
filepaths = [tonga_folder + file for file in tonga_files]
```

```python
def create_month_mapping():

    month_equv = dict()

    for i in range(1, 13):
        month_abbre = datetime.date(1900, i, 1).strftime('%b')
        month_full = datetime.date(1900, i, 1).strftime('%B')
        month_equv.update({month_full: i, month_abbre: i})

    return month_equv


def parse_filename(filename: str):

    filename_lst = filename.replace(".csv", "").split("-")

    identifier = {"year": [], "month": []}
    for elem in filename_lst:
        if elem.isdigit() == True:
            if len(elem) == 4:
                identifier["year"].append(elem)
        else:
            temp_dict = create_month_mapping()
            for (key, val) in temp_dict.items():
                if elem in key:
                    identifier["month"].append(val)
                    
    return identifier
```

```python
tonga = pd.DataFrame()
for file, filepath in zip(tonga_files, filepaths):
    if ".csv" in file:
        print(f"{file} has started.")
        df = pd.read_csv(filepath).drop("Unnamed: 0", axis=1)
        df = df[["Country", "Total"]]
        df["Total"] = (df["Total"].str.replace(",", "")
                       .str.replace("-", "0")
                       .str.replace(" ", "")
                       .astype("int"))
        time = parse_filename(file)
        year, month = time["year"][0], time["month"][0]
        df.columns = ["country", str(year) + "-" + str(month)]
        df["country"] = df["country"].str.replace(" ", "")
        df = df.T
        df.columns = df.iloc[0,].to_list()
        tonga = pd.concat([tonga, df], axis=0)
```

```python
tonga = tonga.drop_duplicates().iloc[1:]
tonga = tonga.reset_index()[tonga.index != "country"]
tonga["index"] = pd.to_datetime(tonga["index"])
tonga["Year"] = tonga["index"].dt.year
tonga = tonga.rename({"index": "Date"}, axis=1)
```
